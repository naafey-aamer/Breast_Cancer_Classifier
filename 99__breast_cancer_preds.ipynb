The Breast Cancer Wisconsin (Diagnostic) Data Set is a well-known dataset commonly used in machine learning and classification tasks. It was originally created by Dr. William H. Wolberg at the University of Wisconsin Hospital to aid in the diagnosis of breast cancer.

This dataset consists of clinical measurements collected from fine needle aspirates (FNA) of breast masses. It includes information about various characteristics extracted from digitized images of the FNA samples, such as the mean, standard error, and worst values of cell nucleus radius, texture, perimeter, area, smoothness, compactness, concavity, concave points, symmetry, and fractal dimension.

Each sample in the dataset is labeled as either benign (non-cancerous) or malignant (cancerous). The goal of using this dataset is often to develop a machine learning model that can accurately predict the diagnosis of breast cancer based on these clinical measurements.

The dataset contains a total of 569 instances, with 30 features (excluding the diagnosis label).

Finding patterns in a dataset of 30 features is difficult for a simple logistic regression model, especially when we are looking for patterns of something as complex as cancer.

I want to see how state of the art we can make this simple sklearn logistic regression model with just feature selection.

For this purpose, I chose to create an evolutionary algorithm to optimize feature selection

```python
size = 6600  #population size
n_feat = 30  # number of features
n_parents = 40  # number of parents
mutation_rate = 0.75  # mutation rate
n_gen = 200  # the number of generations
```

Randomly selected 55% of features for each initial member of the population.

```python
#random initializtion of population
def initilization_of_population(size,n_feat):
    population = []
    for i in range(size):
        chromosome = np.ones(n_feat,dtype=bool)
        chromosome[:int(0.45*n_feat)]=False
        np.random.shuffle(chromosome)
        population.append(chromosome)
    return population
```

```python
#Regression incorprorated fitness score
def fitness_score(population):
    scores = []
    for chromosome in population:
        clf = LogisticRegression()
        clf.fit(X_train.iloc[:, chromosome], y_train)
        predictions = clf.predict(X_test.iloc[:, chromosome])
        score = accuracy_score(y_test, predictions)
        scores.append([score, chromosome])
    scores.sort(key=lambda x: x[0], reverse=True)
    sorted_scores, sorted_population = zip(*scores)
    return sorted_scores, sorted_population
```

Tried the Stochastic Universal Sampling (SUS) method for selection but it offered to significant increase in performance metrics while increasing the computational overhead, hence resorted to simple rank selection.

```python
#Rank method for selection
# def selection(pop_after_fit1, n_parents):
#     population_nextgen = []
#     a = len(pop_after_fit1)
#     for _ in range(n_parents):
#         tournament = random.sample(pop_after_fit1, a)
#         # print(tournament)
#         best_individual1 = max(tournament, key=lambda x: x[0])
#         population_nextgen.append(best_individual1[1])
#     return population_nextgen


#Stochastic Universal Sampling (SUS) method for selection
def selection(pop_after_fit1, n_parents):
    population_nextgen = []

    # Calculate the total fitness of the population
    total_fitness = sum(individual[0] for individual in pop_after_fit1)

    # Calculate the distance between selection points
    distance = total_fitness / n_parents

    # Start with a random offset
    start_offset = random.uniform(0, distance)

    # Initialize the selection pointer
    pointer = start_offset

    # Perform selection
    index = 0
    accumulated_fitness = 0

    for _ in range(n_parents):
        while accumulated_fitness < pointer:
            accumulated_fitness += pop_after_fit1[index][0]
            index = (index + 1) % len(pop_after_fit1)

        population_nextgen.append(pop_after_fit1[index][1])
        pointer += distance

    return population_nextgen
```

```python
def crossover(pop_after_sel):
    population_nextgen = []
    for i in range(len(pop_after_sel)):
        child = pop_after_sel[i]
        child_copy = child.copy()  # Make a copy of the current child
        child_copy[3:12] = pop_after_sel[(i + 1) % len(pop_after_sel)][3:12]
        population_nextgen.append(child_copy)
        population_nextgen.append(child)  # Keep the original child in the next generation as well
    return population_nextgen

# def crossover(pop_after_sel):
#     population_nextgen = []
#     for i in range(len(pop_after_sel)):
#         child = pop_after_sel[i]
#         child_copy = child.copy()  # Make a copy of the current child

#         # Select two random points for crossover
#         point1 = random.randint(0, len(child) - 1)
#         point2 = random.randint(0, len(child) - 1)

#         # Ensure point2 is greater than point1
#         point1, point2 = min(point1, point2), max(point1, point2)

#         # Perform two-point crossover
#         child_copy[point1:point2] = pop_after_sel[(i + 1) % len(pop_after_sel)][point1:point2]

#         population_nextgen.append(child_copy)
#         population_nextgen.append(child)  # Keep the original child in the next generation as well
#     return population_nextgen
```

```python
def mutation(population, mutation_rate):
    mutated_population = []
    for chromosome in population:
        mutated_chromosome = chromosome.copy()
        for i in range(len(mutated_chromosome)):
            if random.random() < mutation_rate:
                mutated_chromosome[i] = not mutated_chromosome[i]
        mutated_population.append(mutated_chromosome)
    return mutated_population
```

The dataset is so small that it converges instantly, hence we use a very high mutation rate to explore the maximum possible combinations of features.

```python
#simulating 200 generations of feature evolution
def generations(size, n_feat, n_parents, mutation_rate, n_gen, X_train, X_test, y_train, y_test):
    best_chromo = None
    best_score = 0
    population_nextgen = initilization_of_population(size, n_feat)
    for i in range(n_gen):
        print("----------------------------------")
        scores, pop_after_fit = fitness_score(population_nextgen)
        print(i, "'s FITTEST SPECIMEN ", scores[0] )
        if scores[0] > best_score and scores[0] != 1.0:
            best_chromo = pop_after_fit[0]
            best_score = scores[0]
        combined_list = list(zip(scores, pop_after_fit))
        pop_after_sel = selection(combined_list, n_parents)
        pop_after_cross = crossover(pop_after_sel)
        population_nextgen = mutation(pop_after_cross, mutation_rate)
        print("BEST_SPECIMEN OVERALL = ", best_score)
    return [best_chromo, best_score]
```

```python
warnings.filterwarnings("ignore")
chromo = generations(size, n_feat, n_parents, mutation_rate, n_gen, X_train, X_test, y_train, y_test)
```

```python
# Get the best individual and its corresponding score
best_individual, best_score = chromo

#Choose the best features
genetically_selected_train = X_train.iloc[:, best_individual]
genetically_selected_test = X_test.iloc[:, best_individual]

#Regression with genetically selected features
logmodel1 = LogisticRegression()
logmodel1.fit(genetically_selected_train, y_train)
predictions = logmodel1.predict(genetically_selected_test)
print("Accuracy score after genetic algorithm is = " + str(accuracy_score(y_test, predictions)))

print("Accuracy score after PCA is = " + str(PCA_accuracy))


#Without Feature Selection
logmodel2 = LogisticRegression()
logmodel2.fit(X_train,y_train)
predictions = logmodel2.predict(X_test)
print("Accuracy without Feature Selection = "+ str(accuracy_score(y_test,predictions)))
```
